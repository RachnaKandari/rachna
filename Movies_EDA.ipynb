```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('movies.csv')
```

```python
df
```

```python
#check for missing data, We are returning a percentage of missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df.boxplot(column=['gross']) #to check for outliers

plt.show()
```

```python
df.drop_duplicates()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=df) #regression plot between gross and budget  to visualize the relationship between them
plt.show()
```

```python
sns.regplot(x="score", y="gross", data=df)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[float, int])
```

```python
numeric_df.corr(method ='pearson')
```

```python
numeric_df.corr(method ='kendall')
```

```python
numeric_df.corr(method ='spearman')
```

```python
correlation_matrix = numeric_df.corr() #by default Pearson

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson') #converting/assigning unique numbers to all the data for corr purpose be it string/categorical value
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5] #high correlation pairs

print(strong_pairs)
```

```python
#Top 15 companies by gross revenue
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')

CompanyGrossSumSorted
```
